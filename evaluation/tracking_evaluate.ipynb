```python
import pandas as pd
import numpy as np

project_name = "Aug_8"

import os
if not os.path.exists(f'../log/{project_name}/control_results'):
    os.makedirs(f'../log/{project_name}/control_results')
```

```python
log = pd.read_csv(f'../log/{project_name}/koopman_results_log.csv')
log
```

```python
franka_results = pd.read_csv(f'../log/{project_name}/franka_control_results.csv')
franka_results['error_eight'] = franka_results['error_eight'].apply(lambda x: float(x[0][0]) if isinstance(x, list) else float(eval(x)[0][0]))
franka_results['error_star'] = franka_results['error_star'].apply(lambda x: float(x[0][0]) if isinstance(x, list) else float(eval(x)[0][0]))
franka_results
```

```python
grouped = franka_results.groupby(['env_name', 'encode_dim', 'use_control_loss', 'use_covariance_loss']).agg({
    'error_eight': ['mean', 'std'],
    'error_star': ['mean', 'std']
}).reset_index()
grouped.to_csv(f'../log/{project_name}/control_results/Franka.csv', index=False)
```

```python
grouped
```

```python
pendulum_results = pd.read_csv(f'../log/{project_name}/pendulum_control_results.csv')
pendulum_results
```

```python
grouped = pendulum_results[['env_name', 'encode_dim', 'use_control_loss', 'use_covariance_loss', 'stability_rate', 'median_best_last10_L1', 'median_best_control_cost']]

# change column name from median_best_last10_L1 to error
grouped = grouped.rename(columns={"median_best_last10_L1": "error", 'median_best_control_cost': 'control_cost'})
grouped[grouped['env_name'] == 'DampingPendulum'].to_csv(f'../log/{project_name}/control_results/DampingPendulum.csv', index=False)
grouped[grouped['env_name'] == 'DoublePendulum'].to_csv(f'../log/{project_name}/control_results/DoublePendulum.csv', index=False)
```

```python
isaacsim_results = pd.read_csv(f'../log/{project_name}/isaac_control_results.csv')
isaacsim_results = isaacsim_results.merge(log[['env_name', 'model_path', 'encode_dim', 'use_control_loss', 'use_covariance_loss', 'seed']], on='model_path', how='left').drop(columns=['model_path', 'robot'])[['env_name','seed', 'encode_dim', 'use_control_loss', 'use_covariance_loss', 'mean_JrPE', 'mean_survival_steps']].sort_values(by=['env_name', 'use_control_loss', 'use_covariance_loss'], ascending=[True, True, True])
isaacsim_results
```

```python
grouped = isaacsim_results.groupby(['env_name', 'encode_dim', 'use_control_loss', 'use_covariance_loss']).agg({
    'mean_JrPE': ['mean', 'std'],
    'mean_survival_steps': ['mean', 'std']
}).reset_index()
grouped[grouped['env_name'] == 'G1'].to_csv(f'../log/{project_name}/control_results/G1.csv', index=False)
grouped[grouped['env_name'] == 'Go2'].to_csv(f'../log/{project_name}/control_results/Go2.csv', index=False)
```

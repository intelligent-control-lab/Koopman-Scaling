```python
import pandas as pd
import numpy as np
```

```python
log = pd.read_csv('../log/Jun_11_control/koopman_results_log.csv')
log
```

```python
franka_results = pd.read_csv('../log/Jun_11_control/franka_control_results.csv')
franka_results['error_eight'] = franka_results['error_eight'].apply(lambda x: float(x[0][0]) if isinstance(x, list) else float(eval(x)[0][0]))
franka_results['error_star'] = franka_results['error_star'].apply(lambda x: float(x[0][0]) if isinstance(x, list) else float(eval(x)[0][0]))
franka_results
```

```python
grouped = franka_results.groupby(['env_name', 'encode_dim', 'use_control_loss', 'use_covariance_loss']).agg({
    'error_eight': 'mean',
    'error_eight': 'std',
    'error_star': 'mean',
    'error_star': 'std'
}).reset_index()
grouped.to_csv('../log/Jun_11_control/control_results/Franka.csv', index=False)
```

```python
pendulum_results = pd.read_csv('../log/Jun_11_control/pendulum_control_results.csv')
pendulum_results
```

```python
temp = pendulum_results[np.isfinite(pendulum_results['control_cost'])]
grouped = temp.groupby(['env_name', 'encode_dim', 'use_control_loss', 'use_covariance_loss']).agg({
    'tracking_error': 'mean',
    'tracking_error': 'std',
    'control_cost': 'mean',
    'control_cost': 'std',
    'seed': 'count'
}).reset_index()
grouped[grouped['env_name'] == 'DampingPendulum'].to_csv('../log/Jun_11_control/control_results/DampingPendulum.csv', index=False)
grouped[grouped['env_name'] == 'DoublePendulum'].to_csv('../log/Jun_11_control/control_results/DoublePendulum.csv', index=False)
```

```python
isaacsim_results = pd.read_csv('../log/Jun_11_control/isaac_control_results.csv')
isaacsim_results = isaacsim_results.merge(log[['env_name', 'model_path', 'encode_dim', 'use_control_loss', 'use_covariance_loss', 'seed']], on='model_path', how='left').drop(columns=['model_path', 'robot'])[['env_name','seed', 'encode_dim', 'use_control_loss', 'use_covariance_loss', 'mean_JrPE', 'time_s']].sort_values(by=['env_name', 'use_control_loss', 'use_covariance_loss'], ascending=[True, True, True])
isaacsim_results
```

```python
grouped = isaacsim_results.groupby(['env_name', 'encode_dim', 'use_control_loss', 'use_covariance_loss']).agg({
    'mean_JrPE': 'mean',
    'mean_JrPE': 'std',
    'time_s': 'mean',
    'time_s': 'std'
}).reset_index()
grouped[grouped['env_name'] == 'G1'].to_csv('../log/Jun_11_control/control_results/G1.csv', index=False)
grouped[grouped['env_name'] == 'Go2'].to_csv('../log/Jun_11_control/control_results/Go2.csv', index=False)
```
